```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=50, centers=2, random_state=4, cluster_std=1)
X += 2
plt.figure(figsize=(5,5))
cm2 = ListedColormap(['#0000ff', '#ff2020'])
cm2 = ListedColormap(['b','r'])
plt.scatter(X[:,0], X[:,1], c=y, cmap=cm2)
plt.xlim(5,15)
plt.ylim(0,10)
plt.show()
```

## Robust Scaler
Sets the median to 0 and caculates the interquartile range

```python
import numpy as np
from sklearn.preprocessing import RobustScaler
sca = RobustScaler()
X_ = sca.fit_transform(X)

plt.figure(figsize=(5,5))
plt.scatter(X_[:,0], X_[:,1], c=y, cmap=cm2)
plt.show()

print(' Before scaling')
print('medians:  {}'.format(sca.center_))
print('IQRs:     {}'.format(sca.scale_))

print(' After scaling')
print('medians:  {}'.format([np.median(X_[:,0]), np.median(X_[:,1])]))
print('IQRs:     {}'.format(np.percentile(X_, 75, axis=0) - np.percentile(X_,25, axis=0)))
```
